2.1 Investigue as proporções de vários tipos de estabelecimentos. Construa um gráfico.

```python
# Proporção por tipo de estabelecimento
type_prop = restLA['object_type'].value_counts(normalize=True).reset_index()
type_prop.columns = ['object_type', 'proportion']
print(type_prop)
```

```python
from plotly import graph_objects as go

object_type = ['Restaurant', 'Fast Food', 'Cafe', 'Pizza', 'Bar', 'Bakery']
proportion = [0.751736, 0.110455, 0.045073, 0.033157, 0.030256, 0.029323]

fig = go.Figure(data=[go.Pie(labels=object_type, values=proportion)])
fig.update_layout(title='Proporção dos Tipos de Estabelecimento em LA')
fig.show()
```

A maior parte proporcionalmente sao restaurantes e fast foods, totalizando 86% do total.
Cafeterias ocupam a terceira posicao.

2.3 Qual tipo de estabelecimento é típico para redes?

```python
# Agrupando por tipo e rede
type_chain = restLA.groupby(['object_type', 'chain']).size().unstack(fill_value=0)

# Proporção dentro de cada tipo
(type_chain.div(type_chain.sum(axis=1), axis=0)).plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Distribuição de Estabelecimentos de Rede por Tipo')
plt.ylabel('Proporção')
plt.xlabel('Tipos de Estabelecimentos')
plt.tight_layout()
plt.show()
```

As confeitarias, pizzarias, fast food e cafeterias são estabelecimentos típicos para rede.

2.5 Determine o número médio de assentos para cada tipo de restaurante. Em média, qual tipo de restaurante tem o maior número de assentos? Construa gráficos.
